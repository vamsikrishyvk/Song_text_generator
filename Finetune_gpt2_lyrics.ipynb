```python
import glob

parent_dir = '/content/data/Albums/'
file_pattern = '*.txt'
search_pattern = parent_dir + '**/' + file_pattern
file_paths = glob.glob(search_pattern, recursive=True)

```

```python
file_paths
```

```python
file_paths[444:445]
```

```python
import pandas as pd
import glob


data = []

for path in file_paths:

    with open(path, 'r') as file:
        file_data = file.read()

    data.append(file_data)
```

```python
train_data  = data[0:444]
test_data   = data[444:]
```

```python
test_data[0]
```

```python
df_train = pd.DataFrame(train_data, columns=['Text'])
df_test  = pd.DataFrame(test_data, columns=['Text'])

import os 

if not os.path.exists('content'):
    os.makedirs('content')
  
pd.DataFrame({"lyrics": df_train['Text']})\
    .to_csv(os.path.join('content', 'lyrics_train.csv'), index=False)

pd.DataFrame({"lyrics": df_test['Text']})\
    .to_csv(os.path.join('content', 'lyrics_test.csv'), index=False)
```

```python
for model_name in ["124M"]:  # Choose from ["124M","355M","774M"]
    gpt2.download_gpt2(model_name=model_name)   # model is saved into current directory under /models/124M/
 
```

```python
learning_rate = 0.0001
optimizer = 'adam' # adam or sgd
batch_size = 1
model_name = "124M" # Has to match one downloaded locally
```

```python
import sys
import csv
maxInt = sys.maxsize

while True:
    # decrease the maxInt value by factor 10 
    # as long as the OverflowError occurs.

    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
import tensorflow as tf
tf.compat.v1.reset_default_graph()
sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
              '/content/content/content/lyrics_train.csv',
              model_name=model_name,
              sample_every=50,
              save_every=50,
              print_every=10,
              learning_rate=learning_rate,
              batch_size=batch_size,
              restore_from='latest',
              steps=1000)   # max number of training steps
```
